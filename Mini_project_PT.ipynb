## Tahap 1: Pembersihan Data (Preprocessing)

### Langkah 1: Menambahkan Header (Nama Kolom)

```python
import pandas as pd

df = pd.read_csv('data_mentah.csv', header=None)
df.columns = [
    "conversation_id_str",
    "created_at",
    "favorite_count",
    "full_text",
    "id_str",
    "image_url",
    "in_reply_to_screen_name",
    "lang",
    "location",
    "quote_count",
    "reply_count",
    "retweet_count",
    "tweet_url",
    "user_id_str",
    "username"
]

print("Data setelah ditambahkan header:")
print(df.head())

df.to_csv('dataset.csv', index=False)

print("\nFile berhasil disimpan sebagai 'dataset.csv'")
```

### Langkah 2: Case Folding

```python
import pandas as pd

df = pd.read_csv('dataset.csv')

print("Sebelum case folding:")
print(df['full_text'].head())

# --- Tahap Case Folding ---
df['full_text'] = df['full_text'].str.lower()

print("\nSetelah case folding:")
print(df['full_text'].head())

df.to_csv('data_casefolding.csv', index=False)
print("\nHasil disimpan ke 'data_casefolding.csv'")
```

### Langkah 3: Cleaning (Pembershan Teks)

```python
import pandas as pd
import re

df = pd.read_csv('data_casefolding.csv')

print("=== Sebelum Cleaning ===")
display(df['full_text'].head(15))

def clean_text(text):
    text = re.sub(r'http\S+', '', text)          # hapus URL
    text = re.sub(r'@\w+', '', text)             # hapus mention
    text = re.sub(r'#', '', text)                # hapus tanda #
    text = re.sub(r'[^a-zA-Z\s]', '', text)      # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text).strip()     # hapus spasi ganda
    return text

df['clean_text'] = df['full_text'].apply(clean_text)

print("\n=== Setelah Cleaning ===")
display(df[['full_text', 'clean_text']].head(15))
```

### Langkah 4: Tokeniznig

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

print("=== Sebelum Tokenizing ===")
display(df['clean_text'].head(15))

df['tokens'] = df['clean_text'].apply(word_tokenize)

print("\n=== Setelah Tokenizing ===")
display(df[['clean_text', 'tokens']].head(15))
```

### Langkah 5: Text Filtering

```python
!pip install Sastrawi
```

```python
import pandas as pd
import nltk
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from tqdm import tqdm

factory = StopWordRemoverFactory()
stopwords_id = set(factory.get_stop_words())

extra_stopwords = {'yg', 'aja', 'nih', 'dong', 'sih', 'ya', 'deh', 'lah', 'wkwk', 'hehe', 'hihi', 'haha', 'ampun'}
stopwords_id.update(extra_stopwords)

tqdm.pandas()

def clean_and_filter(tokens):
    cleaned_tokens = []
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF" 
        u"\U0001F1E0-\U0001F1FF"  
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )

    for word in tokens:
        word = emoji_pattern.sub(r'', word) 
        if word and word not in stopwords_id:
            cleaned_tokens.append(word)

    return cleaned_tokens

print("=== Sebelum Filtering (Stopword + Emoji Removal) ===")
display(df['tokens'].head(15))

df['filtered_tokens'] = df['tokens'].progress_apply(clean_and_filter)

print("\n=== Setelah Filtering (Stopword + Emoji Removal) ===")
display(df[['tokens', 'filtered_tokens']].head(15))
```

### Langkah 6: Stemming (Text Parsing)

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tqdm import tqdm

tqdm.pandas()

factory = StemmerFactory()
stemmer = factory.create_stemmer()

print("=== Sebelum Stemming (Text Parsing) ===")
display(df['filtered_tokens'].head(30))

def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

df['stemmed_tokens'] = df['filtered_tokens'].progress_apply(stem_tokens)

print("\n=== Setelah Stemming (Text Parsing) ===")
display(df[['filtered_tokens', 'stemmed_tokens']].head(30))
```
