## Langkah 3: Case Folding

```python
import pandas as pd

df = pd.read_csv('data_bersih.csv')

if df.columns[0] != 'full_text':
    df.columns = ['full_text']

print("Sebelum case folding:")
print(df['full_text'].head(10))

df['full_text'] = df['full_text'].astype(str).str.lower()

print("\nSetelah case folding:")
print(df['full_text'].head(10))

df.to_csv('data_casefolding.csv', index=False)
print("\nHasil disimpan ke 'data_casefolding.csv'")
```

## Langkah 4: Cleaning (Pembershan Teks)

```python
import pandas as pd
import re
import string

df = pd.read_csv('data_casefolding.csv')

print("=== Sebelum Cleaning ===")
display(df['full_text'].head(15))

def clean_text(text):
    text = str(text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#(\w+)', r'\1', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_text'] = df['full_text'].apply(clean_text)

print("\n=== Setelah Cleaning ===")
display(df['clean_text'].head(10))

df[['clean_text']].to_csv('data_cleaning.csv', index=False)

print("\nFile 'data_cleaning.csv' berhasil disimpan (hanya kolom clean_text)")
```

## Langkah 6: Text Filtering

```python
import pandas as pd
import nltk
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from tqdm import tqdm
import ast

df = pd.read_csv('data_tokenizing.csv')
df['full_text'] = df['full_text'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

print("=== Sebelum Filtering (Stopword + Emoji Removal) ===")
display(df['full_text'].head(10))

factory = StopWordRemoverFactory()
stopwords_id = set(factory.get_stop_words())

extra_stopwords = {
    'yg', 'aja', 'nih', 'dong', 'sih', 'ya', 'deh', 'lah', 
    'wkwk', 'hehe', 'hihi', 'haha', 'ampun', 'eh', 'mah', 
    'loh', 'lho', 'kok', 'gitu', 'sama', 'buat', 'kayak'
}
stopwords_id.update(extra_stopwords)
tqdm.pandas()

def clean_and_filter(tokens):
    cleaned_tokens = []
    emoji_pattern = re.compile(
        "[" 
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF" 
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )

    for word in tokens:
        word = emoji_pattern.sub(r'', word)
        if word and word not in stopwords_id and len(word) > 1:
            cleaned_tokens.append(word)
    
    return cleaned_tokens

df['full_text'] = df['full_text'].progress_apply(clean_and_filter)

print("\n=== Setelah Filtering (Stopword + Emoji Removal) ===")
display(df['full_text'].head(10))
df.to_csv('data_filtering.csv', index=False)

print("\n File 'data_filtering.csv' berhasil disimpan (kolom full_text berisi token hasil filtering).")
```

## Langkah 7: Stemming (Text Parsing)

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tqdm import tqdm
import ast

df = pd.read_csv('data_filtering.csv')
df['full_text'] = df['full_text'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

print("=== Sebelum Stemming (Text Parsing) ===")
display(df['full_text'].head(10))

factory = StemmerFactory()
stemmer = factory.create_stemmer()

tqdm.pandas()
def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

df['full_text'] = df['full_text'].progress_apply(stem_tokens)

print("\n=== Setelah Stemming (Text Parsing) ===")
display(df['full_text'].head(10))

df.to_csv('data_stemming.csv', index=False)

print("\n File 'data_stemming.csv' berhasil disimpan (kolom full_text berisi token hasil stemming).")
```

## Langkah 1: Normalisasi Kata

```python
import pandas as pd
from tqdm import tqdm
import ast
import re

df = pd.read_csv('data_stemming.csv')
df['full_text'] = df['full_text'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

print("=== Sebelum Normalisasi ===")
display(df['full_text'].head(10))

normalization_dict = {
    'gk': 'tidak', 'ga': 'tidak', 'nggak': 'tidak', 'ngga': 'tidak', 'tdk': 'tidak',
    'tak': 'tidak', 'enggak': 'tidak', 'kagak': 'tidak', 'gak': 'tidak',
    'bgt': 'banget', 'bngt': 'banget',
    'yg': 'yang', 'yng': 'yang',
    'aja': 'saja', 'aj': 'saja',
    'trs': 'terus', 'trus': 'terus',
    'sm': 'sama', 'dgn': 'dengan', 'dg': 'dengan',
    'utk': 'untuk', 'buat': 'untuk',
    'krn': 'karena', 'karen': 'karena', 'karna': 'karena',
    'dlm': 'dalam', 'jd': 'jadi', 'jdi': 'jadi',
    'sdh': 'sudah', 'udh': 'sudah', 'udah': 'sudah', 'suda': 'sudah',
    'blm': 'belum', 'blum': 'belum',
    'pd': 'pada', 'tp': 'tapi', 'tpi': 'tapi',
    'klo': 'kalau', 'klu': 'kalau', 'klw': 'kalau',
    'dpt': 'dapat', 'dptlah': 'dapat',
    'dgnmu': 'denganmu', 'smua': 'semua',
    'mkn': 'makan', 'mknya': 'makannya',
    'tlg': 'tolong', 'tolongin': 'tolong',
    'bru': 'baru', 'baru2': 'baru',
    'org': 'orang', 'org2': 'orang', 'orgnya': 'orangnya',
    'jg': 'juga', 'jga': 'juga', 'jugaa': 'juga',
    'sy': 'saya', 'aq': 'aku', 'gw': 'saya', 'gue': 'saya', 'gua': 'saya',
    'km': 'kamu', 'kmu': 'kamu', 'lo': 'kamu', 'lu': 'kamu', 'loe': 'kamu',
    'trs2': 'terus', 'trus2': 'terus',
    'ok': 'oke', 'okee': 'oke', 'okey': 'oke', 'sip': 'oke',
    'pls': 'tolong', 'plis': 'tolong', 'please': 'tolong',
    'btw': 'ngomongngomong', 'imo': 'menurutku', 'idk': 'tidaktahu',
    'mantul': 'mantap', 'mantab': 'mantap', 'mantep': 'mantap',
    'makasih': 'terimakasih', 'thx': 'terimakasih', 'thanks': 'terimakasih'
}

def normalize_tokens(tokens):
    normalized = []
    for word in tokens:
        word = re.sub(r'(.)\1{2,}', r'\1', word)
        word = normalization_dict.get(word, word)
        if len(word) < 3:
            continue
        if word:
            normalized.append(word)

    return normalized

tqdm.pandas()
df['full_text'] = df['full_text'].progress_apply(normalize_tokens)

print("\n=== Setelah Normalisasi ===")
display(df['full_text'].head(10))

df.to_csv('data_normalisasi.csv', index=False)
print("\n File 'data_normalisasi.csv' berhasil disimpan (kolom full_text sudah dinormalisasi).")
```

## Langkah 2: Bi-gram

```python
import pandas as pd
from tqdm import tqdm
import ast
from nltk import ngrams

df = pd.read_csv('data_normalisasi.csv')
df['full_text'] = df['full_text'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

print("=== Contoh Data Sebelum Bigram ===")
display(df['full_text'].head(15))
def make_bigrams(tokens):
    return [' '.join(gram) for gram in ngrams(tokens, 2)] if len(tokens) >= 2 else []

tqdm.pandas()
df['bigrams'] = df['full_text'].progress_apply(make_bigrams)

print("\n=== Contoh Data Setelah Bigram ===")
display(df[['full_text', 'bigrams']].head(15))

df.to_csv('data_bigram.csv', index=False)
print("\n File 'data_bigram.csv' berhasil disimpan (berisi kolom full_text dan bigrams).")
```

## Langkah 3: Tri-gram

```python
import pandas as pd
from tqdm import tqdm
import ast
from nltk import ngrams

df = pd.read_csv('data_bigram.csv')
df['full_text'] = df['full_text'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

print("=== Contoh Data Sebelum Trigram ===")
display(df['full_text'].head(5))

def make_trigrams(tokens):
    return [' '.join(gram) for gram in ngrams(tokens, 3)] if len(tokens) >= 3 else []

tqdm.pandas()
df['trigrams'] = df['full_text'].progress_apply(make_trigrams)

print("\n=== Contoh Data Setelah Trigram ===")
display(df[['full_text', 'trigrams']].head(5))

df.to_csv('data_trigram.csv', index=False)

print("\n File 'data_trigram.csv' berhasil disimpan (berisi kolom full_text dan trigrams).")
```

## Langkah 4: Worldcloud

```python
import pandas as pd
import ast
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from tqdm import tqdm
import os

tqdm.pandas()

INPUT_FILES = ['data_trigram.csv', 'data_bigram.csv', 'data_normalisasi.csv']
NGRAM_COL_PREFER = ['trigrams', 'bigrams', 'full_text']
OUTPUT_IMAGE = 'wordcloud_trigram.png'
MAX_WORDS = 300
FIGSIZE = (12, 8) 
BG_COLOR = 'white'

df = None
for fn in INPUT_FILES:
    if os.path.exists(fn):
        df = pd.read_csv(fn)
        print(f" Membaca file: {fn}")
        break

if df is None:
    raise FileNotFoundError(f"Tidak ditemukan file input. Pastikan salah satu dari {INPUT_FILES} ada di working directory.")
ngram_col = None
for c in NGRAM_COL_PREFER:
    if c in df.columns:
        ngram_col = c
        break

if ngram_col is None:
    if len(df.columns) == 1:
        ngram_col = df.columns[0]
    else:
        raise KeyError(f"Tidak menemukan kolom n-gram. Diharapkan satu dari {NGRAM_COL_PREFER} ada di file.")

print(f" Menggunakan kolom: '{ngram_col}' untuk membuat word cloud")

def ensure_list(x):
    if isinstance(x, list):
        return x
    if pd.isna(x):
        return []
    if isinstance(x, str):
        try:
            val = ast.literal_eval(x)
            return val if isinstance(val, list) else [str(x)]
        except Exception:
            return x.split()
    return [x]

df[ngram_col] = df[ngram_col].progress_apply(ensure_list)

all_ngrams = []
for tokens in df[ngram_col]:
    for t in tokens:
        if isinstance(t, (list, tuple)):
            all_ngrams.append(' '.join(t))
        else:
            all_ngrams.append(str(t).strip())

freq = Counter(all_ngrams)
if len(freq) == 0:
    raise ValueError("Tidak ada n-gram ditemukan di data. Periksa isi kolom input.")
print("\nðŸ“Š 20 n-gram teratas:")
for i, (term, count) in enumerate(freq.most_common(20), start=1):
    print(f"{i:2d}. {term} â€” {count}")
wc = WordCloud(
    width=1600,
    height=900,
    background_color=BG_COLOR,
    max_words=MAX_WORDS,
    collocations=False
).generate_from_frequencies(freq)

plt.figure(figsize=FIGSIZE)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
plt.show()

wc.to_file(OUTPUT_IMAGE)
print(f"\n Word cloud disimpan ke: {OUTPUT_IMAGE}")
```
